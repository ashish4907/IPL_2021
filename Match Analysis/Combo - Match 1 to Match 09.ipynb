# BATSMAN

```python
df.head(1)
```

```python
batsman_played= df[['Team','Match Played','Batsman Runs','Balls','Fours','Sixes','Strike Rate','50s','100s','Vs','Times Out']]
```

```python
Batsman= batsman_played[batsman_played['Balls'] >=1]
```

```python
Batsman.replace(to_replace=['Yes','No'],value=[0,1],inplace=True)
```

```python
Batsman['Times Out'] = Batsman['Times Out'].astype(float)
```

```python
BR= Batsman.groupby(by=['Players','Team'])[['Match Played','Batsman Runs','Balls','Fours','Sixes','50s','100s','Times Out']].sum().nlargest(100,columns=['Batsman Runs'])
```

```python
BR['Strike Rate'] = BR['Batsman Runs']/BR['Balls']*100
```

```python
batsman= BR.round(2)
```

```python
batsman.loc[batsman['Times Out'] ==0,'Average']  = batsman['Batsman Runs']/batsman['Match Played']
batsman.loc[(batsman['Times Out'] == 1) & (batsman['Match Played'] == 1),'Average'] = batsman['Batsman Runs']
batsman.loc[(batsman['Times Out'] == 1) & (batsman['Match Played'] >= 2),'Average'] = batsman['Batsman Runs']/(batsman['Match Played']-batsman['Times Out'])
batsman.loc[(batsman['Times Out'] == 2) & (batsman['Match Played'] == 2),'Average'] = batsman['Batsman Runs']
```

```python
batsman.loc[batsman['Balls'] >=1 , 'Innings'] = batsman['Match Played']
```

```python
batsman.head(1)
```

```python
batsman[batsman['Times Out'] ==1]
```

# BOWLERS

```python
bowler_played= df[['Team','Match Played','Over','Bowler Balls','Maiden','Bowler Runs','Wickets','Economy Rate','Vs']]
```

```python
Bowler= bowler_played[bowler_played['Bowler Balls'] >=1]
```

```python
W= Bowler.groupby(by=['Players','Team'])[['Match Played','Over','Bowler Balls','Maiden','Bowler Runs','Wickets']].sum().nlargest(100,'Wickets')
```

```python
W['Economy'] = W['Bowler Runs']/W['Over']
```

```python
W['Average'] = W['Bowler Runs']/W['Wickets']
```

```python
W['Strike Rate'] = W['Bowler Balls']/W['Wickets']
```

```python
bowler = W.round(2)
```

```python
bowler.sort_values(by='Wickets',ascending=False).shape
```

## Most Wickets

```python
bowler[['Match Played','Over','Wickets','Average']].nlargest(10,'Wickets')
```

## Best Bowling Average

```python
bowler[['Match Played','Over','Bowler Runs','Wickets','Average']].nsmallest(10,'Average')
```

## Best Economy

```python
bowler[['Match Played','Over','Wickets','Economy']].nsmallest(10,'Economy')
```

## Best Strike Rate

```python
SR= bowler[bowler['Wickets'] >=5].nlargest(5,'Strike Rate')
```

```python
SR[['Match Played','Over','Wickets','Strike Rate']].nsmallest(5,'Strike Rate')
```
