```python
import pandas as pd
import os
```

```python
files= [file for file in os.listdir('./CSV Files')]

all_matches_data=pd.DataFrame()

for file in files:
    df=pd.read_csv('./CSV Files/'+file)
    all_matches_data=pd.concat([all_matches_data,df])
```

```python
all_matches_data.to_csv('Combo_Match1_to_Match5.csv',index=False)
```

```python
 df= pd.read_csv('Combo CSV Files\Combo_Match1_to_Match5.csv',index_col='Players')
```

```python
df.head(1)
```

```python
batsman_played = df[['Team','Match Played','Batsman Runs','Balls','Fours','Sixes','Strike Rate','50s','100s']]
```

```python
Batsman= batsman_played[batsman_played['Balls'] >=1]
```

```python
Batsman.head(1)
```

```python
BR= Batsman.groupby('Players')[['Match Played','Batsman Runs','Balls','Fours','Sixes','50s','100s']].sum().nlargest(100,columns=['Batsman Runs'])
```

```python
BR.head(1)
```

```python
BR['Strike Rate']= (BR['Batsman Runs']/BR['Balls'])*100
```

```python
BR['Average'] = BR['Batsman Runs']/BR['Match Played']
```

```python
batsman=BR.round(2)
```

```python
batsman.head(1)
```

```python
batsman[['Match Played','Fours']].nlargest(10,'Fours').head(1)
```

```python
batsman[(batsman['Batsman Runs'] >=50)].nlargest(10,columns='Strike Rate').head(1)
```
